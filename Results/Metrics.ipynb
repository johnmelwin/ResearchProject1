```python
import matplotlib.pyplot as plt
import numpy as np

# Model names
models = ['GPT-3.5', 'Fine-tuned GPT-3.5 (100-NoVAL)', 'Fine-tuned GPT-3.5 (100)', 'Fine-tuned GPT-3.5 (5000)']

# F1 Scores for each model
f1_scores = [0.50, 0.63, 0.72,  0.82]

# Setting up the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Base x locations for the bars
index = np.arange(len(models))

# Bar width
bar_width = 0.5

# Plotting F1 scores
bars = ax.bar(index, f1_scores, bar_width, color='blue', label='F1 Score')

# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Model')
ax.set_ylabel('F1 Score')
ax.set_title('F1 Score Comparison Across Models')
ax.set_xticks(index)
ax.set_xticklabels(models)
ax.legend()

# Attach a text label above each bar, displaying its height
def autolabel(bars):
    """Attach a text label above each bar in *bars*, displaying its height."""
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(round(height, 2)),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(bars)

plt.tight_layout()
plt.show()
```


In the realm of software development, the ability to accurately trace requirements to their corresponding source code is a critical aspect of quality assurance and project management. This traceability ensures that every feature or requirement outlined in the project documentation is implemented correctly in the codebase. The classification reports provided offer a detailed comparison between the performance of the original GPT-3.5 model and a fine-tuned version of GPT-3.5 in accomplishing this task.

The initial classification report for the original GPT-3.5 model showcases its capabilities in identifying whether a requirement is linked to the source code. With a precision and recall of 0.91 and 0.50, respectively, for the 'False' and 'True' classes, the model demonstrates a solid foundation in distinguishing non-linked cases. However, the same level of accuracy does not extend to identifying linked requirements. A precision and recall of 0.50 for the 'True' class indicates a balanced but moderate performance. This signifies that while the model can identify linked requirements to some extent, it does so with a high rate of false negatives. In practical terms, this means the model often fails to recognize when a requirement is indeed traceable to source code, which can lead to oversight in ensuring that all project requirements are properly implemented.

Understanding the critical nature of reducing false negatives in this contextwhere missing a link between a requirement and its implementation could lead to significant project delays or failuresthe fine-tuning process was undertaken. The objective was clear: enhance the model's ability to identify true links between requirements and source code without significantly compromising its ability to correctly identify non-linked cases.

The classification report for the fine-tuned GPT-3.5 model reveals a marked improvement in its performance. Precision for the 'True' class increased to 0.76, and recall to 0.89, indicating a substantial reduction in false negatives. This improvement suggests that the fine-tuned model is significantly better at correctly identifying requirements that are linked to source code, thus ensuring a more reliable traceability process. Meanwhile, the 'False' class saw an increase in precision to 0.98 and a slight decrease in recall to 0.95, indicating that while the model became better at identifying non-linked cases, it did so with a negligible increase in false positives.

The overall accuracy of the model improved from 85% to 94%, reflecting the success of the fine-tuning process in enhancing the model's traceability capabilities. This improvement in macro and weighted averages across precision, recall, and the F1-score for both classes underscores the fine-tuned model's balanced performance across the board.

In summary, the fine-tuning of the GPT-3.5 model has resulted in a significant enhancement of its ability to trace requirements to source code. By reducing false negatives, the model now provides a more reliable tool for software developers and project managers in ensuring that all project requirements are accurately implemented in the codebase. This advancement not only improves the quality assurance process but also contributes to the overall success and efficiency of software development projects.
